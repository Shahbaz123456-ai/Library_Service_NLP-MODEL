```python
import pandas as pd
data = pd.read_csv('Library_Services_20240831.csv')
data
```

```python
data.isnull().sum()
```

```python
# Display relevant text columns
data[['Branch', 'Location']].head()
```

```python
import re
import string

# Function to clean text data
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)  # Remove punctuation
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.strip()  # Remove whitespaces
    return text

# Apply cleaning function to 'Location' column
data['cleaned_location'] = data['Location'].apply(clean_text)

# Display cleaned text
data[['cleaned_location']].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize vectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features=100)

# Apply vectorization on the cleaned text
X = vectorizer.fit_transform(data['cleaned_location'])

# Display feature names (words)
print(vectorizer.get_feature_names_out())
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'Branch' is the target label (for classification), we encode it
data['Branch_encoded'] = data['Branch'].astype('category').cat.codes

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, data['Branch_encoded'], test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize the classifier
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
```

```python
# Convert the 'Branch' column to a categorical type
data['Branch'] = data['Branch'].astype('category')

# Example of predicting a branch based on location text
new_location = ["11701 New Hampshire Avenue Silver Spring"]
new_location_cleaned = [clean_text(loc) for loc in new_location]
new_location_vectorized = vectorizer.transform(new_location_cleaned)

# Predict the branch
predicted_branch_code = model.predict(new_location_vectorized)
predicted_branch = data['Branch'].cat.categories[predicted_branch_code[0]]

print(f"Predicted branch: {predicted_branch}")
```
